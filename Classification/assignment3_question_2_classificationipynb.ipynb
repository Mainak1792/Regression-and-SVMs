# Classification using Gaussian (RBF) kernel SVM

```python
data = dataset.iloc[:,1:16]
labels = dataset.iloc[:, 0]
print(data)
```

- Classification using Gaussian (RBF) kernel SVM: k(x_1, x_2) = exp (-||(x_1 - x_2||^2/2 \sigma^2))

- C - hyperplane/reguarisation parameter
- Gamma - (1/2sigma^2)

Show a matrix of your cross validation results, where the entry (i, j) of the matrix corresponds to the classification accuracy on the cross validation set with ith value of C and j th value of \sigma. Next, use the entire training set to train an SVM classifier with the best C and \sigma values determined via the cross validation procedure outlined above.

```python
score1=[]
svm10_score=[]
svm50_score=[]
svm100_score=[]
svm175_score=[]
svm250_score=[]


Cm = [1, 10, 50,100,175, 250, 300,500]
# 1e-1, 1, 1e1
# gammam = [0.001,0.01,0.1,1,10]
gammam = [1e-2, 1e-1,0.01,1, 10, 100, 1e2,1e1]
classifiers = []



from sklearn.model_selection import KFold
cv = KFold(n_splits=5)
for train_index, test_index in cv.split(data):


    X_train, X_test, y_train, y_test = data.iloc[train_index], data.iloc[test_index], labels.iloc[train_index], labels.iloc[test_index]
    X_train = X_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
    X_test = X_test.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)

    
    for C in Cm:
      for gamma in gammam:
        clf = SVC(C=C, gamma=gamma)
        clf.fit(X_train, y_train)
        score1.append(clf.score(X_test, y_test))
        classifiers.append((C, gamma, np.mean(score1)))

    
    # clf12.fit(X_train, y_train)
    # clf1.fit(X_train, y_train)
    


    # svml_score.append(clf12.score(X_test, y_test))
    # svmr_score.append(clf1.score(X_test, y_test))
    
#

# print("svm", svml_score)
# print(np.mean(svml_score))
# print("svmr", svmr_score)
# print(np.mean(svmr_score))
```
